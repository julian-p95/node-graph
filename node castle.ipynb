```python
import networkx as nx
import matplotlib.pyplot as plt
# creating an empty graph
G = nx.Graph()

# adding nodes to the graph
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G'])

# adding edges with weights to the graph
G.add_edge('A', 'B', weight=10)
G.add_edge('A', 'F', weight=9)
G.add_edge('B', 'C', weight=5)
G.add_edge('B', 'D', weight=8)
G.add_edge('B', 'E', weight=13)
G.add_edge('C', 'G', weight=15)
G.add_edge('F', 'D', weight=5)
G.add_edge('F', 'G', weight=3)
G.add_edge('D', 'C', weight=4)
G.add_edge('E', 'D', weight=2)
```

```python
# defining the positions of the nodes using the spring layout algorithm
pos = nx.spring_layout(G)

# drawing the graph with labels and edge weights
nx.draw(G, pos, with_labels=True)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')


# displaying the graph
plt.show()
```
